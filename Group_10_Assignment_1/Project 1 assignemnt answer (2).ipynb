## Pollution Levels Analysis by Region

```python
# Check available columns to identify region information
print("Available columns in the dataset:")
print(data.columns.tolist())
print("\nFirst few rows:")
data.head()
```

```python
data["Address"].unique()
```

```python
# Extract region/district from Address column
# Pattern: "number, street, District-gu, Seoul, Republic of Korea"
# Extract the district name (e.g., "Jongno-gu", "Jung-gu", etc.)

def extract_region(address):
    parts = address.split(', ')
    # The district is typically the 3rd element (index 2)
    if len(parts) >= 3:
        return parts[2]  # This will be the "-gu" district
    return None

data['region'] = data['Address'].apply(extract_region)
print("Extracted regions:")
print(data['region'].unique())
print(f"\nNumber of unique regions: {data['region'].nunique()}")
```

```python
data.columns
```

```python
data.head()
```

```python
# Group by region and calculate average pollution levels
pollution_by_region = data.groupby('region')[['PM2.5', 'PM10', 'O3', 'NO2', 'SO2', 'CO']].mean()
print("Average Pollution Levels by Region:")
print(pollution_by_region.round(2))
```

```python
pollution_by_region.T
```

```python
# Bar chart: Average PM2.5 levels by region
plt.figure(figsize=(14, 6))
pollution_by_region['PM2.5'].sort_values(ascending=False).plot(kind='bar', color='steelblue')
plt.title('Average PM2.5 Levels by Region', fontsize=14, fontweight='bold')
plt.xlabel('Region (District)', fontsize=12)
plt.ylabel('PM2.5 Level (g/m)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Bar chart: Average PM10 levels by region
plt.figure(figsize=(14, 6))
pollution_by_region['PM10'].sort_values(ascending=False).plot(kind='bar', color='coral')
plt.title('Average PM10 Levels by Region', fontsize=14, fontweight='bold')
plt.xlabel('Region (District)', fontsize=12)
plt.ylabel('PM10 Level (g/m)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Comparison: Multiple pollutants by region (top 10 regions with highest PM2.5)
top_regions = pollution_by_region['PM2.5'].nlargest(10).index
comparison_data = pollution_by_region.loc[top_regions, ['PM2.5', 'PM10', 'O3', 'NO2']]

plt.figure(figsize=(14, 7))
comparison_data.plot(kind='bar', width=0.8)
plt.title('Top 10 Most Polluted Regions - Multi-Pollutant Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Region (District)', fontsize=12)
plt.ylabel('Pollution Level (g/m)', fontsize=12)
plt.legend(title='Pollutant Type', loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
